```python
import numpy as np
import imageio.v2 as iio
from matplotlib.colors import rgb_to_hsv, hsv_to_rgb
import matplotlib.pyplot as plt

# Đọc ảnh RGB và chuẩn hóa về [0,1]
img = iio.imread('bird.png')  # Đổi tên file nếu cần
img_rgb = img[..., :3] / 255.0

# Chuyển sang HSV
img_hsv = rgb_to_hsv(img_rgb)

# Thay đổi kênh H và V
img_hsv[..., 0] = img_hsv[..., 0] / 3      # Hnew = 1/3 * Hold
img_hsv[..., 2] = img_hsv[..., 2] * 0.75   # Vnew = 3/4 * Vold

# Chuyển lại sang RGB
img_new = hsv_to_rgb(img_hsv)
img_new_uint8 = (img_new * 255).astype(np.uint8)

# Lưu ảnh mới
iio.imwrite('hsv_modified.png', img_new_uint8)

# Hiển thị ảnh gốc và ảnh đã chỉnh sửa
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(img_rgb)
axs[0].set_title('Ảnh gốc')
axs[0].axis('off')
axs[1].imshow(img_new)
axs[1].set_title('Ảnh HSV đã chỉnh sửa')
axs[1].axis('off')
plt.show()
```
