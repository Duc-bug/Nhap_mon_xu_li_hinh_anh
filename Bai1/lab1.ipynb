```python
import numpy as np
import imageio.v2 as iio
import matplotlib.pylab as plt

# c nh
data = iio.imread('bird.png')

# To nh ch c knh 
red_img = np.zeros_like(data)
red_img[..., 0] = data[..., 0]
iio.imwrite('bird_red.png', red_img)

# To nh ch c knh xanh l
green_img = np.zeros_like(data)
green_img[..., 1] = data[..., 1]
iio.imwrite('bird_green.png', green_img)

# To nh ch c knh xanh dng
blue_img = np.zeros_like(data)
blue_img[..., 2] = data[..., 2]
iio.imwrite('bird_blue.png', blue_img)

# Hin th nh
fig, axs = plt.subplots(1, 3, figsize=(12,4))
axs[0].imshow(red_img)
axs[0].set_title('Red Channel')
axs[1].imshow(green_img)
axs[1].set_title('Green Channel')
axs[2].imshow(blue_img)
axs[2].set_title('Blue Channel')
for ax in axs:
    ax.axis('off')
plt.show()
```
