```python
import numpy as np
import imageio.v2 as iio
import matplotlib.pylab as plt
#đọc ảnh
img = iio.imread('bird.png') 
#Hoán đổi màu
img_rb = img[..., [2, 1, 0]]  
iio.imwrite('swap_rb.png', img_rb)
img_rg = img[..., [1, 0, 2]]  
iio.imwrite('swap_rg.png', img_rg)
img_gb = img[..., [0, 2, 1]] 
iio.imwrite('swap_gb.png', img_gb)
#Hiện thị ảnh
fig, axs = plt.subplots(1, 3, figsize=(12, 4))
axs[0].imshow(img_rb)
axs[0].set_title('Red <-> Blue')
axs[1].imshow(img_rg)
axs[1].set_title('Red <-> Green')
axs[2].imshow(img_gb)
axs[2].set_title('Green <-> Blue')
for ax in axs:
    ax.axis('off')
plt.show()
```
